```python
import os

def update_yolo_labels(dir1, dir2):
    """
    Update YOLO class numbers in .txt files for two directories.

    Class 0 -> 3

    Class 1 -> 1
    Class 2 -> 1
    Class 3 -> 1

    Class 4 -> 0
    Class 5 -> 2

    Class 7 -> 4
    Class 8 -> 4
    Class 9 -> 4
    Class 10 -> 4
    Class 11 -> 4
    Class 12 -> 4

    Class 13 -> 5
    Class 14 -> 5
    Class 15 -> 5

    Args:
        dir1 (str): Path to the first directory containing .txt files.
        dir2 (str): Path to the second directory containing .txt files.
    """

    def update_labels_in_dir(folder_path):
        for filename in os.listdir(folder_path):
            if filename.endswith(".txt"):
                file_path = os.path.join(folder_path, filename)

                # Read the file content
                with open(file_path, 'r') as file:
                    lines = file.readlines()

                # Update the class numbers
                updated_lines = []
                for line in lines:
                    parts = line.strip().split(' ')
                    class_number = int(parts[0])
                    if class_number == 0:
                        parts[0] = '3'  # Change class 0 to 3
                    elif class_number in [1, 2, 3]:
                        parts[0] = '1'  # Change class 1, 2, or 3 to 1
                    elif class_number == 4:
                        parts[0] = '0'  # Change class 4 to 0
                    elif class_number == 5:
                        parts[0] = '2'  # Change class 5 to 2
                    elif class_number in [7, 8, 9, 10, 11, 12]:
                        parts[0] = '4'  # Change class 7, 8, 9, 10, 11, or 12 to 4
                    elif class_number in [13, 14, 15]:
                        parts[0] = '5'  # Change class 13, 14, or 15 to 5

                    updated_lines.append(' '.join(parts))

                # Write the updated content back to the file
                with open(file_path, 'w') as file:
                    file.write('\n'.join(updated_lines))

    # Update the labels in both directories
    update_labels_in_dir(dir1)
    update_labels_in_dir(dir2)

    print("Class number updates completed for both directories!")

# Example usage:
# update_yolo_labels('path_to_first_directory', 'path_to_second_directory')

# Directories containing the label files for training and validation
train_label_dir = "/content/drive/MyDrive/__Dandoon__/dataset_6classes_Ultimate/labels/train"
val_label_dir = "/content/drive/MyDrive/__Dandoon__/dataset_6classes_Ultimate/labels/val"

update_yolo_labels(train_label_dir, val_label_dir)
```

```python
import os

def load_ground_truth_labels(image_path):
    # Replace the 'images' directory with 'labels' and change the extension to .txt
    base_label_path = image_path.replace('images', 'labels')

    # Handle both .jpg and .png cases
    if base_label_path.endswith('.jpg'):
        label_path = base_label_path.rsplit('.jpg', 1)[0] + '.txt'
    elif base_label_path.endswith('.png'):
        label_path = base_label_path.rsplit('.png', 1)[0] + '.txt'
    else:
        label_path = None  # If the file extension is neither .jpg nor .png

    boxes = []
    if label_path and os.path.exists(label_path):
        with open(label_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                parts = line.strip().split()
                # The label format in YOLO is <class> <x_center> <y_center> <width> <height>
                class_id = int(parts[0])
                x_center, y_center, width, height = map(float, parts[1:])
                boxes.append((class_id, x_center, y_center, width, height))
    return boxes
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
import os

model_path = '/content/drive/MyDrive/results_6class_Ultimate/100epochs/detect/custom_yolov8/weights/best.pt'

# Load your trained model
model_best = YOLO(model_path)

# Directory containing test images and corresponding ground truth labels
image_dir = '/content/drive/MyDrive/__Dandoon__/dataset_6classes_Ultimate/images/val'
save_dir = '/content/drive/MyDrive/dataset_6classes_Ultimate/100epochs/pred_img_30cof'  # Directory to save results
os.makedirs(save_dir, exist_ok=True)  # Create the directory if it doesn't exist

# Run inference on all images in the directory
results = model_best(image_dir, conf=0.3)

# def load_ground_truth_labels(image_path):
#     # Replace the 'images' directory with 'labels' and change the extension to .txt
#     base_label_path = image_path.replace('images', 'labels')

#     # Handle both .jpg and .png cases
#     if base_label_path.endswith('.jpg'):
#         label_path = base_label_path.replace('.jpg', '.txt')
#     elif base_label_path.endswith('.png'):
#         label_path = base_label_path.replace('.png', '.txt')
#     else:
#         label_path = None  # If the file extension is neither .jpg nor .png

#     boxes = []
#     if label_path and os.path.exists(label_path):
#         with open(label_path, 'r') as f:
#             lines = f.readlines()
#             for line in lines:
#                 parts = line.strip().split()
#                 # The label format in YOLO is <class> <x_center> <y_center> <width> <height>
#                 class_id = int(parts[0])
#                 x_center, y_center, width, height = map(float, parts[1:])
#                 boxes.append((class_id, x_center, y_center, width, height))
#     return boxes

# Visualize predictions and ground truths
for i, result in enumerate(results):
    # Load image and corresponding ground truth labels
    img_path = result.path  # Path to the image
    img = cv2.imread(img_path)
    ground_truth_boxes = load_ground_truth_labels(img_path)

    # Plot ground truth boxes (in green)
    for gt_box in ground_truth_boxes:
        class_id, x_center, y_center, width, height = gt_box
        h, w, _ = img.shape
        # Convert YOLO format to corner format for drawing
        x1 = int((x_center - width / 2) * w)
        y1 = int((y_center - height / 2) * h)
        x2 = int((x_center + width / 2) * w)
        y2 = int((y_center + height / 2) * h)
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box for ground truth
        cv2.putText(img, f'GT: {class_id}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Plot predicted boxes (in red)
    for pred_box in result.boxes:
        x1, y1, x2, y2 = map(int, pred_box.xyxy[0])  # Get predicted box coordinates
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)  # Red box for prediction
        class_id = int(pred_box.cls[0])
        conf = pred_box.conf[0]
        cv2.putText(img, f'Pred: {class_id} ({conf:.2f})', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

    # Save the image with both boxes
    save_path = os.path.join(save_dir, os.path.basename(img_path))  # Save with the same image name
    cv2.imwrite(save_path, img)
    print(f"Saved image with predictions and ground truth: {save_path}")
```
